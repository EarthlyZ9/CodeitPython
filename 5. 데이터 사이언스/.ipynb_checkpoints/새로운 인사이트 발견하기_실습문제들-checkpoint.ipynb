### 박물관이 살아 있다 1  
시설명에 '대학'이 포함되어 있으면 '대학', 그렇지 않으면 '일반'으로 나누어 '분류' column에 입력한다.

```python
df_museum = pd.read_csv('/Users/ljs2000/데이터사이언스/csv/museum_1.csv')
df_museum
```

```python
is_univ = df_museum['시설명'].str.contains('대학')
is_univ
```

```python
df_museum.loc[is_univ == True, '분류'] = '대학'
df_museum.loc[is_univ == False, '분류'] = '일반'
df_museum
```

### 박물관이 살아있다 2  
솔희가 박물관의 위치를 파악할 수 있게 '운영기관전화번호' column의 맨 앞 3자리를 추출하고, '지역번호' column에 넣어주세요

```python
df_museum2 = pd.read_csv('/Users/ljs2000/데이터사이언스/csv/museum_2.csv')
df_museum2
```

```python
phone_number = df_museum2['운영기관전화번호'].str.split(pat = '-', expand = True)
phone_number
```

```python
df_museum2['지역번호'] = phone_number[0]
df_museum2
```

### 박물관이 살아있다3  
'지역번호' column을 '지역명' 으로 변경하고, 아래 규칙에 따라 지역을 넣어주세요.

```python
df_museum3 = df_museum2
df_museum3
```

```python
local_number ={
    '02':'서울시',
    '031':'경기도',
    '032':'경기도',
    '033':'강원도',
    '041':'충청도',
    '042':'충청도',
    '043':'충청도',
    '044':'충청도',
    '051':'부산시',
    '052':'경상도',
    '053':'경상도',
    '054':'경상도',
    '055':'경상도',
    '061':'전라도',
    '062':'전라도',
    '063':'전라도',
    '064':'제주도',
    '1577':'기타',
    '070':'기타'
}
```

```python
df_museum3['지역번호'] = df_museum3['지역번호'].map(local_number)
```

```python
df_museum3.rename(columns={'지역번호':'지역명'}, inplace = True)
df_museum3
```
