```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/ljs2000/PycharmProjects/CodeitPython/5. 데이터 사이언스/csv/enrolment_3.csv')
df
```

```python
allowed = df['status'] == 'allowed'
course_counts = df.loc[allowed,'course name'].value_counts()
course_counts
```

```python
auditorium_list = list(course_counts[course_counts >= 80].index)
largeroom_list = list(course_counts[(course_counts < 80) & (course_counts >= 40)].index)
mediumroom_list = list(course_counts[(course_counts < 40) & (course_counts >= 15)].index)
smallroom_list = list(course_counts[(course_counts < 15) & (course_counts >= 5)].index)
```

```python
for i in range(len(auditorium_list)):
    df.loc[(df['course name'] == sorted(auditorium_list)[i]) & allowed, 'room assignment'] = 'Auditorium-' + str(i+1)
```

```python
for i in range(len(largeroom_list)):
    df.loc[(df['course name'] == sorted(largeroom_list)[i]) & allowed, 'room assignment'] = 'Large-' + str(i+1)
```

```python
for i in range(len(mediumroom_list)):
    df.loc[(df['course name'] == sorted(mediumroom_list)[i]) & allowed, 'room assignment'] = 'Medium-' + str(i+1)
```

```python
for i in range(len(smallroom_list)):
    df.loc[(df['course name'] == sorted(smallroom_list)[i]) & allowed, 'room assignment'] = 'Small-' + str(i+1)
```

```python
df.rename(columns = {'room assignment': 'room number'}, inplace = True)
df
```
