```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/ljs2000/PycharmProjects/CodeitPython/5. 데이터 사이언스/csv/enrolment_2.csv')
df
```

```python
df['room assignment'] = 'not assigned'
```

```python
student_counts = df['course name'].value_counts()
```

```python
auditorium_course = list(student_counts[student_counts >= 80].index)
for course_name in auditorium_course:
    df.loc[df['course name'] == course_name, 'room assignment'] = 'Auditorium'
```

```python
large_room_condition = student_counts[student_counts >= 40] & student_counts[student_counts < 80]
large_room_course = list(student_counts[large_room_condition].index)
for course_name in large_room_course:
    df.loc[df['course name'] == course_name, 'room assignment'] = 'Large room'
```

```python
medium_room_condition = student_counts[student_counts >= 15] & student_counts[student_counts < 40]
medium_room_course = list(student_counts[medium_room_condition].index)
for course_name in medium_room_course:
    df.loc[df['course name'] == course_name, 'room assignment'] = 'Medium room'
```

```python
small_room_condition = student_counts[student_counts >= 5] & student_counts[student_counts < 15]
small_room_course = list(student_counts[small_room_condition].index)
for course_name in small_room_course:
    df.loc[df['course name'] == course_name, 'room assignment'] = 'Small room'
```
